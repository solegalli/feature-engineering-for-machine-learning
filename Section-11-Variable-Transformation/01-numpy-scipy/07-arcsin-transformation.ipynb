```python
import numpy as np
import pandas as pd

# for plotting
import matplotlib.pyplot as plt

# for Q-Q plots
import scipy.stats as stats

# dataset for the demo
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
#Load dataset
breast_cancer = load_breast_cancer()
X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
y = breast_cancer.target

X.head()
```

```python
# Separate data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
variables = [
  'mean compactness',
  'mean concavity',
  'mean concave points',
  'mean fractal dimension',
  'smoothness error',
  'compactness error',
  'concavity error',
  'concave points error',
  'symmetry error',
  'fractal dimension error',
  'worst symmetry',
  'worst fractal dimension']
```

```python
# make a copy of the data

train_t = X_train.copy()
test_t = X_test.copy()
```

```python
# transform the data
train_t[variables] = np.arcsin(np.sqrt(X_train[variables]))
test_t[variables] = np.arcsin(np.sqrt(X_test[variables]))
```

```python
# original variables
X_train[variables].hist(figsize=(20,20), bins=30)
plt.show()
```

```python
# transformed variable
train_t[variables].hist(figsize=(20,20), bins=30)
plt.show()
```
